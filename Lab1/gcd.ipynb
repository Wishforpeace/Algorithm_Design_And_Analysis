```python
import time
import matplotlib.pyplot as plt
def euclidean_gcd(m, n):
    while n != 0:
        m, n = n, m % n
    return m
```

```python
input_sizes = list(range(1, 1001))  # 输入规模从1到1000
runtimes = []

for size in input_sizes:
    start_time = time.time()
    euclidean_gcd(size, size // 2)  # 在这里输入m和n的值，这里以m为size，n为size // 2为例
    end_time = time.time()
    runtimes.append(end_time - start_time)
```

```python
plt.plot(input_sizes, runtimes, label="Euclidean GCD")
plt.xlabel("Input Size")
plt.ylabel("Runtime (seconds)")
plt.legend()
plt.title("Runtime vs. Input Size for GCD Algorithm")
plt.show()
```

```python
import time
import matplotlib.pyplot as plt

# 线性时间复杂度算法
def linear_gcd(m, n):
    while n != 0:
        m, n = n, m % n
    return m

# 平方根时间复杂度算法
def sqrt_gcd(m, n):
    while n != 0:
        m, n = n, m % n
    return m

# 对数时间复杂度算法
def log_gcd(m, n):
    while n != 0:
        m, n = n, m % n
    return m

# 测试函数执行时间
def measure_execution_time(func, *args):
    start_time = time.time()
    result = func(*args)
    end_time = time.time()
    execution_time = end_time - start_time
    return result, execution_time

# 主实验函数
def experiment(max_value):
    x = []
    y_linear = []
    y_sqrt = []
    y_log = []

    for i in range(1, max_value + 1):
        result, execution_time = measure_execution_time(linear_gcd, i, max_value)
        x.append(i)
        y_linear.append(execution_time)

        result, execution_time = measure_execution_time(sqrt_gcd, i, max_value)
        y_sqrt.append(execution_time)

        result, execution_time = measure_execution_time(log_gcd, i, max_value)
        y_log.append(execution_time)

    return x, y_linear, y_sqrt, y_log

# 设置最大输入值
max_value = 100

# 执行实验
x, y_linear, y_sqrt, y_log = experiment(max_value)

# 绘制图表
plt.plot(x, y_linear, label="Linear Time")
plt.plot(x, y_sqrt, label="Square Root Time")
plt.plot(x, y_log, label="Logarithmic Time")
plt.xlabel("min(m, n)")
plt.ylabel("Time (s)")
plt.title("Time Complexity Comparison")
plt.legend()
plt.show()
```
